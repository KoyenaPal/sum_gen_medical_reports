```python
import pandas as pd
import csv
from google.colab import drive
drive.mount('/content/gdrive', force_remount=False)
root_dir = "/content/gdrive/My Drive/masters_thesis/"
```

```python
def create_training_testing(data_file, training_frac, training_csv, testing_csv):
  fieldnames = ["hadm_id", "source", "target"]
  df = pd.read_csv(data_file)
  split_val = int(len(df) * training_frac)
  training_df = df.iloc[:split_val,]
  testing_df = df.iloc[split_val:,]
  with open(training_csv, 'w') as csvfile1:
    csvwriter = csv.DictWriter(csvfile1, delimiter=',', fieldnames=fieldnames)
    csvwriter.writerow(dict((fn,fn) for fn in fieldnames))
    test_array = []
    for index, row in training_df.iterrows():
      source = row["source"]
      target = row["target"]
      hadm_id = row["hadm_id"]
      test_array.append({"source": source, "target": target, "hadm_id": hadm_id})
    for row in test_array:
      csvwriter.writerow(row)
  with open(testing_csv, 'w') as csvfile2:
    csvwriter2 = csv.DictWriter(csvfile2, delimiter=',', fieldnames=fieldnames)
    csvwriter2.writerow(dict((fn,fn) for fn in fieldnames))
    test_array = []
    for index, row in testing_df.iterrows():
      source = row["source"]
      target = row["discharge_summary"]
      hadm_id = row["hadm_id"]
      test_array.append({"source": source, "target": target, "hadm_id": hadm_id})
    for row in test_array:
      csvwriter2.writerow(row)


source1 = root_dir + "setup1_nursing_discharge.csv"
training1 = root_dir + "setup1_training.csv"
testing1 = root_dir + "setup1_testing.csv"
source2 = root_dir + "setup2_luhn_discharge.csv"
training2 = root_dir + "setup2_training.csv"
testing2 = root_dir + "setup2_testing.csv"
source3 = root_dir + "setup3_sentence_3_summary.csv"
training3 = root_dir + "setup3_training.csv"
testing3 = root_dir + "setup3_testing.csv"
source4 = root_dir + "setup4_nursing_hist_section.csv"
training4 = root_dir + "setup4_training.csv"
testing4 = root_dir + "setup4_testing.csv"
source5 = root_dir + "setup5_nursing_hist_discharge.csv"
training5 = root_dir + "setup5_training.csv"
testing5 = root_dir + "setup5_testing.csv"

training_frac = 0.7
#create_training_testing(source1, training_frac, training1, testing1)
#create_training_testing(source2, training_frac, training2, testing2)
#create_training_testing(source3, training_frac, training3, testing3)
#create_training_testing(source4, training_frac, training4, testing4)
#create_training_testing(source5, training_frac, training5, testing5)
```
